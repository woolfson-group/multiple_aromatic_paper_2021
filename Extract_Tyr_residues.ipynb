```python
import Bio.PDB as PDB
import os
import itertools

#This script inputs the directory of coiled coil regions pdb files and outputs a directory of pdb files 
##containing only Tyr residues. Biological units may not contain Tyr residues or a CC region,
##which results in empty pdb files.

p = PDB.PDBParser()
parent_directory  = "./"
pdb_file_directory = parent_directory+"Complete_biological_units_CC/"
save_file_directory = parent_directory+"Complete_biological_units_CC_Tyr/"

class TyrSelect(PDB.Select):
    def accept_residue(self, residue):
        if residue.get_resname()=='TYR':
            return 1
        else:
            return 0
```

```python
for pdb_entry in os.listdir(pdb_file_directory):
    if pdb_entry[-3:] != "pdb":
        print(pdb_entry)
        pass
    else:
        structure = p.get_structure('X', pdb_file_directory+pdb_entry)
        io = PDB.PDBIO()
        io.set_structure(structure)
        io.save(save_file_directory+pdb_entry[:-4]+'_TYR.pdb', TyrSelect())
```
