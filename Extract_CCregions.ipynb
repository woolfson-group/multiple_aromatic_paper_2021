```python
import Bio.PDB as PDB
import os
import itertools
import json

##This script inputs the coiled coil JSON dictionary from SOCKET_extract-CCregions.ipynb and the corresponding PDB
##files and outputs a directory of coiled coil regions pdb files. Biological units may not contain the CC region 
##which results in empty pdb files.

p = PDB.PDBParser()
parent_directory  = "./"
pdb_file_directory = parent_directory+"Complete_biological_units/"
save_file_directory = parent_directory+"Complete_biological_units_CC/"
cc_region_dict_directory = parent_directory+"CCregions2020.json"

with open(cc_region_dict_directory) as json_file: 
    allpdb_cc_regions = json.load(json_file) 

class TyrSelect(PDB.Select):
    def accept_residue(self, residue):
        cc_residues = allpdb_cc_regions[residue.full_id[0][0:4]]
        for cc_regions in cc_residues:
            if residue.full_id[2] == cc_regions[0]:
                if residue.full_id[3][1] in list(range(int(cc_regions[1]),int(cc_regions[2]))):
                    return 1
        else:
            return 0
```

```python
for pdb_entry in os.listdir(pdb_file_directory):
    if (#pdb_entry[-4:] == "pdb1" or pdb_entry[-4:] == "pdb2" or pdb_entry[-4:] == "pdb3" or pdb_entry[-4:] == "pdb4" or
    pdb_entry[-4:] == "pdb5" or pdb_entry[-4:] == "pdb6" or pdb_entry[-4:] == "pdb7" or pdb_entry[-4:] == "pdb8"):
        structure = p.get_structure(pdb_entry, pdb_file_directory+pdb_entry)
        io = PDB.PDBIO()
        io.set_structure(structure)
        io.save(save_file_directory+pdb_entry[:-5]+'_CCbiolunit'+pdb_entry[-1:]+'.pdb', TyrSelect())
    else:
        print(pdb_entry)
        pass
```
